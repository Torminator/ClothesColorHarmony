```python
import sys
sys.path.append('../')
from colorharmony.munsell import Munsell
import matplotlib.pyplot as plt
import matplotlib.patches as patches
%matplotlib inline
```

```python
munsell_colors = list(Munsell().values())
munsell = Munsell()
munsell.revert_mapping()
```

```python
hue1 = '5Y'
chroma_value_hue1 = [color for color in munsell_colors if color[0] == hue1]
hue2 = '5GY'
chroma_value_hue2 = [color for color in munsell_colors if color[0] == hue2]
```

```python
import matplotlib.patches as patches

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111)

for hue, value, chroma in chroma_value_hue1:
    color = [channel/255 for channel in munsell[(hue, value, chroma)]]
    ax.add_patch(patches.Rectangle((chroma/2, value), width=1, height=1, color=color))

for hue, value, chroma in chroma_value_hue2:
    color = [channel/255 for channel in munsell[(hue, value, chroma)]]
    ax.add_patch(patches.Rectangle((-chroma/2, value), width=1, height=1, color=color))

for i in range(0,11):
    ax.add_patch(patches.Rectangle((0, i), width=1, height=1, color=(i/10, i/10, i/10)))

ax.set_facecolor((0.5, 0.5, 0.5))
ax.set_title("{} - {}".format(hue2, hue1))
ax.set_xlabel("Chroma")
ax.set_ylabel("Value")

plt.xlim((-8, 8))
plt.ylim((-1, 12))
```

```python
#colors = [('5Y', 6+i, 8-2*i) for i in range(3)]
colors = [('5Y', 6, 8), ('5Y', 5, 2)]

fig = plt.figure(figsize=(16,10))
ax = fig.add_subplot(111)

for i, color in enumerate(colors):
    color_rgb = [channel/255 for channel in munsell[color]]
    ax.add_patch(patches.Rectangle((i, 0), width=1, height=1, color=color_rgb))

ax.set_facecolor((0.5, 0.5, 0.5))
ax.set_title("Compare colors")

plt.xlim((0, len(colors)))
plt.ylim((0, 1))
```
